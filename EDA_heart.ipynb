**Heart Attack Analysis & Prediction Dataset**

About this dataset

Age : Age of the patient
Sex : Sex of the patient
exang: exercise induced angina (1 = yes; 0 = no)
ca: number of major vessels (0-3)
cp : Chest Pain type chest pain type

Value 1: typical angina
Value 2: atypical angina
Value 3: non-anginal pain
Value 4: asymptomatic
trtbps : resting blood pressure (in mm Hg)

chol : cholestoral in mg/dl fetched via BMI sensor
fbs : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
rest_ecg : resting electrocardiographic results

Value 0: normal
Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
thalach : maximum heart rate achieved

target : 0= less chance of heart attack 1= more chance of heart attack

**Removal of Outlier**

```python
import scipy       # used to making data normally distributed.
from scipy import stats
from scipy.stats import zscore
from scipy.stats.mstats import winsorize
```

# z-score will help us to understand how much it is deviated from the mean. z-scor = (mean - observation)/standard devoation


```python
zscore_trtbps = zscore(df["RestingBP"])
for threshold in range(1,4):
    print("Threshold Value : {}".format(threshold))
    print("Number of Outliers : {}".format(len(np.where(zscore_trtbps > threshold)[0])))
    # indicate how many outlier are above the given threshold
    print("\n")
```

```python
df[zscore_trtbps > 2][['RestingBP']]
```

```python
df[zscore_trtbps > 2][['RestingBP']].RestingBP.min() # minimum value of outlier
```

```python
df[df['RestingBP']<170].RestingBP.max() # indicate the max datapoint after which outlier found
```

```python
winsorize_percentile_trtbps = (stats.percentileofscore(df['RestingBP'],165)/100)
print(winsorize_percentile_trtbps)
```

```python
1 - winsorize_percentile_trtbps
```

```python
trtbps_winsorize = winsorize(df.RestingBP, (0, (1 - winsorize_percentile_trtbps)));
```

```python
plt.boxplot(trtbps_winsorize)
plt.xlabel("trtbps_winsorise", color = 'g')
plt.show()
```

```python
df["trtbps_winsorize"] = trtbps_winsorize
```
